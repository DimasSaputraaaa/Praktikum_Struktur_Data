## Struktur Graf dengan adjacency List

```python
# Membuat graf tak berarah menggunakan adjacency List
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Menampikan graf
def print_graph(graf):
  for node in graf:
    print(f"{node} -> {graf[node]}")

print_graph(graph)
```

## Traversal graf menggunakan BFS

```python
from collections import deque
def bfs(graph, start):
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=" ")
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

# Contoh pemanggilan
bfs(graph, 'A')

```

## Traversal graf menggunakan DFS

```python
def dfs(graph, start, visited=None):
  if visited is None:
    visited = set()

  if start not in visited:
    print(start, end=" ")
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)

# Contoh Pemangglan
dfs(graph, 'A')
```

## Latihan/ Tugas

```python
# ============================================================
#  Graf Tak Berarah — BFS, DFS, find_path, is_connected
# ============================================================

from collections import deque


# ----------------------------------------------------------
# 1. REPRESENTASI GRAF (adjacency list)
# ----------------------------------------------------------

def create_graph():
    """Membuat graf tak berarah dengan 6 simpul dan 8 sisi."""
    graph = {
        'A': ['B', 'C', 'D'],
        'B': ['A', 'C', 'E'],
        'C': ['A', 'B', 'F'],
        'D': ['A', 'E'],
        'E': ['B', 'D', 'F'],
        'F': ['C', 'E'],
    }
    return graph


def print_graph(graph):
    """Mencetak representasi adjacency list dari graf."""
    print("=" * 40)
    print("      GRAF TAK BERARAH")
    print("=" * 40)

    # Hitung jumlah sisi (tiap sisi terhitung 2x di adj-list)
    total_edges = sum(len(neighbors) for neighbors in graph.values()) // 2

    print(f"  Jumlah simpul : {len(graph)}")
    print(f"  Jumlah sisi   : {total_edges}")
    print("-" * 40)
    print("  Adjacency List:")
    for node in sorted(graph):
        neighbors = " → ".join(sorted(graph[node]))
        print(f"    {node}  :  {neighbors}")

    print("-" * 40)
    print("  Visualisasi sisi:")
    printed = set()
    for node in sorted(graph):
        for neighbor in sorted(graph[node]):
            edge = tuple(sorted((node, neighbor)))
            if edge not in printed:
                print(f"    {edge[0]} ──── {edge[1]}")
                printed.add(edge)
    print("=" * 40)


# ----------------------------------------------------------
# 2. BFS — Breadth-First Search (cetak + kembalikan list)
# ----------------------------------------------------------

def bfs(graph, start):
    """
    BFS dari simpul `start`.
    Mencetak urutan kunjungan dan mengembalikan list-nya.
    """
    visited = set()
    queue   = deque([start])
    order   = []               # ← Task 3: list urutan kunjungan

    visited.add(start)

    while queue:
        node = queue.popleft()
        order.append(node)

        for neighbor in sorted(graph[node]):   # sorted agar deterministik
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    print(f"\n[BFS dari '{start}']")
    print("  Urutan kunjungan:", " → ".join(order))
    return order          # Task 3: kembalikan list


# ----------------------------------------------------------
# 2. DFS — Depth-First Search (rekursif)
# ----------------------------------------------------------

def dfs(graph, start, visited=None, order=None):
    """
    DFS rekursif dari simpul `start`.
    Panggilan pertama: dfs(graph, 'A')
    """
    if visited is None:
        visited = set()
        order   = []

    visited.add(start)
    order.append(start)

    for neighbor in sorted(graph[start]):
        if neighbor not in visited:
            dfs(graph, neighbor, visited, order)

    return order


def run_dfs(graph, start):
    """Wrapper DFS yang juga mencetak hasilnya."""
    order = dfs(graph, start)
    print(f"\n[DFS dari '{start}']")
    print("  Urutan kunjungan:", " → ".join(order))
    return order


# ----------------------------------------------------------
# 4. find_path — DFS untuk menemukan satu jalur start → end
# ----------------------------------------------------------

def find_path(graph, start, end, visited=None, path=None):
    """
    Mengembalikan satu jalur dari `start` ke `end` (list simpul),
    atau None jika tidak ada jalur.
    Menggunakan DFS secara rekursif.
    """
    if visited is None:
        visited = set()
        path    = []

    visited.add(start)
    path = path + [start]          # buat salinan agar backtrack aman

    if start == end:
        return path

    for neighbor in sorted(graph[start]):
        if neighbor not in visited:
            result = find_path(graph, neighbor, end, visited, path)
            if result is not None:
                return result

    return None          # tidak ditemukan jalur melalui simpul ini


# ----------------------------------------------------------
# 5. is_connected — cek konektivitas seluruh graf
# ----------------------------------------------------------

def is_connected(graph):
    """
    Mengembalikan True jika semua simpul dalam graf saling terhubung,
    False jika terdapat simpul yang terisolasi dari yang lain.
    Strategi: jalankan DFS dari simpul pertama;
    jika semua simpul terkunjungi → terhubung.
    """
    if not graph:
        return True

    start      = next(iter(graph))          # simpul pertama (sembarang)
    reachable  = set(dfs(graph, start))     # simpul yang bisa dicapai
    all_nodes  = set(graph.keys())

    return reachable == all_nodes


# ============================================================
#  MAIN — demonstrasi semua fungsi
# ============================================================

if __name__ == "__main__":

    # ── 1. Buat & tampilkan graf ──────────────────────────
    graph = create_graph()
    print_graph(graph)

    # ── 2 & 3. BFS (cetak + kembalikan list) ─────────────
    bfs_order = bfs(graph, 'A')
    print(f"  (list dikembalikan BFS): {bfs_order}")

    # ── 2. DFS ───────────────────────────────────────────
    dfs_order = run_dfs(graph, 'A')
    print(f"  (list dikembalikan DFS): {dfs_order}")

    # ── 4. find_path ─────────────────────────────────────
    print("\n[find_path]")
    pairs = [('A', 'F'), ('D', 'C'), ('F', 'D')]
    for s, e in pairs:
        path = find_path(graph, s, e)
        if path:
            print(f"  Jalur {s} → {e}: {' → '.join(path)}")
        else:
            print(f"  Jalur {s} → {e}: tidak ditemukan")

    # ── 5. is_connected ───────────────────────────────────
    print("\n[is_connected]")
    connected = is_connected(graph)
    print(f"  Graf terhubung? {'YA ✓' if connected else 'TIDAK ✗'}")

    # Uji dengan graf tidak terhubung
    disconnected_graph = {
        'A': ['B'],
        'B': ['A'],
        'C': ['D'],   # komponen terpisah
        'D': ['C'],
    }
    result2 = is_connected(disconnected_graph)
    print(f"  Graf terputus (A-B  |  C-D) terhubung? "
          f"{'YA ✓' if result2 else 'TIDAK ✗'}")
```
