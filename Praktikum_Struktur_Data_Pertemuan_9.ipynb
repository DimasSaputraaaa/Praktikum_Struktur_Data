## IMPLEMENTASI HASH TABLE SEDERHANA

```python
class SimpleHashTable:
  def __init__(self, size):
    self.size = size
    self.table = [None] * size

  def hash_function(self, key):
    return hash(key) % self.size

  def insert(self, key, value):
    index = self.hash_function(key)
    self.table[index] = value

  def search(self, key):
    index = self.hash_function(key)
    return self.table[index]

# contoh penggunaan
ht = SimpleHashTable(10)
ht.insert("apel", 100)
ht.insert("pisang", 200)

print("nilai untuk 'apel':", ht.search("apel"))
print("nilai untuk 'pisang':", ht.search("pisang"))
```

## PENANGANAN COLLISION DENGAN LINIER PROBING

```python
class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        original_index = index

        while self.table[index] is not None:
            # jika key sudah ada
            if self.table[index][0] == key:
                break

            index = (index + 1) % self.size

            if index == original_index:
                raise Exception("Hash table penuh")

        self.table[index] = (key, value)

    def search(self, key):
        index = self.hash_function(key)
        original_index = index

        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]

            index = (index + 1) % self.size

            if index == original_index:
                break

        return None


# contoh penggunaan
ht = LinearProbingHashTable(5)

ht.insert("apel", 100)
ht.insert("pisang", 200)
ht.insert("melon", 300)

print("nilai untuk 'apel':", ht.search("apel"))
print("nilai untuk 'melon':", ht.search("melon"))
```

## PENANGANAN COLLISION DENGAN CHAINING

```python
class ChainingHashTable:
  def __init__(self, size):
    self.size = size
    self.table = [[] for _ in range(size) ]

  def hash_function (self, key) :
    return hash (key) % self.size

  def insert(self, key, value):
    index = self.hash_function (key)
    for i, (k, v) in enumerate(self.table[index]) :
      if k == key:
       self.table[index] [i] = (key, value)
       return
    self.table[index] .append( (key, value) )

  def search(self, key):
    index = self.hash_function (key)
    for k, v in self. table[index] :
      if k == key:
        return v
    return None

# Contoh penggunaan
ht = ChainingHashTable(5)

ht.insert ("apel", 100)
ht.insert ("pisang", 200)
ht.insert ("melon", 300)

print ("Nilai untuk 'apel':", ht.search ("apel"))
```

## LATIHAN/ TUGAS

```python
# ============================================================
#   EKSPERIMEN HASH TABLE: Linear Probing & Chaining
# ============================================================

# ─────────────────────────────────────────────────────────────
# EKSPERIMEN 1 — Linear Probing Hash Table
# ─────────────────────────────────────────────────────────────

class LinearProbingHashTable:
    def __init__(self, size):
        self.size  = size
        self.table = [None] * size          # Slot kosong = None

    # --- Fungsi hash dasar: ord(char) % size ---
    def hash_function(self, key):
        return ord(key[0]) % self.size

    def insert(self, key, value):
        index     = self.hash_function(key)
        original  = index
        probe_num = 0

        print(f"\n  Menyisipkan '{key}': {value}")
        print(f"    hash('{key}') = ord('{key[0]}') % {self.size} "
              f"= {ord(key[0])} % {self.size} = {index}")

        while self.table[index] is not None:
            existing_key = self.table[index][0]
            # Update jika key sama
            if existing_key == key:
                self.table[index] = (key, value)
                print(f"    → Update di slot {index}")
                return
            # Collision → coba slot berikutnya
            probe_num += 1
            print(f"    ⚠ Collision di slot {index} "
                  f"(ditempati '{existing_key}') → probe ke-{probe_num}")
            index = (original + probe_num) % self.size
            if index == original:
                raise Exception("Tabel penuh! Tidak bisa menyisipkan data.")

        self.table[index] = (key, value)
        print(f"    ✓ Disimpan di slot {index}")

    def display(self):
        print("\n" + "=" * 45)
        print("  Isi Tabel (Linear Probing)")
        print("=" * 45)
        print(f"  {'Slot':<6} {'Key':<8} {'Value'}")
        print("  " + "-" * 25)
        for i, entry in enumerate(self.table):
            if entry is None:
                print(f"  [{i}]    {'—':<8} {'—'}")
            else:
                print(f"  [{i}]    {entry[0]:<8} {entry[1]}")
        print("=" * 45)


# ─────────────────────────────────────────────────────────────
# EKSPERIMEN 2 — Chaining Hash Table
# ─────────────────────────────────────────────────────────────

class ChainingHashTable:
    def __init__(self, size):
        self.size  = size
        self.table = [[] for _ in range(size)]  # Setiap slot = list (chain)

    def hash_function(self, key):
        # Jumlah ord setiap karakter, modulo size
        return sum(ord(c) for c in key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        hash_detail = " + ".join(f"ord('{c}')={ord(c)}" for c in key)
        total = sum(ord(c) for c in key)

        print(f"\n  Menyisipkan '{key}': {value}")
        print(f"    hash('{key}') = ({hash_detail})")
        print(f"               = {total} % {self.size} = {index}")

        # Cek apakah key sudah ada (update)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                print(f"    → Update di slot {index}")
                return

        # Collision jika slot sudah ada isi
        if self.table[index]:
            existing = [k for k, _ in self.table[index]]
            print(f"    ⚠ Collision di slot {index} "
                  f"(sudah ada: {existing}) → ditambahkan ke chain")
        else:
            print(f"    ✓ Disimpan di slot {index} (slot kosong)")

        self.table[index].append((key, value))

    def display(self):
        print("\n" + "=" * 50)
        print("  Isi Tabel (Chaining)")
        print("=" * 50)
        for i, chain in enumerate(self.table):
            if not chain:
                print(f"  Slot [{i}] → (kosong)")
            else:
                items = " → ".join(f"('{k}', {v})" for k, v in chain)
                print(f"  Slot [{i}] → {items}")
        print("=" * 50)


# ─────────────────────────────────────────────────────────────
# MAIN — Jalankan kedua eksperimen
# ─────────────────────────────────────────────────────────────

def main():
    # ── Eksperimen 1: Linear Probing ──────────────────────────
    print("\n" + "█" * 50)
    print("  EKSPERIMEN 1: LINEAR PROBING (ukuran tabel = 5)")
    print("█" * 50)

    lp = LinearProbingHashTable(size=5)

    data_lp = [("A", 10), ("B", 20), ("C", 30), ("D", 40), ("E", 50)]
    for key, value in data_lp:
        lp.insert(key, value)

    lp.display()

    print("""
  📝 PENJELASAN — Apa yang terjadi jika dua key hash-nya sama?
  ─────────────────────────────────────────────────────────────
  Misalkan key "A" dan key "F" keduanya menghasilkan hash = 0.

  Langkah-langkah Linear Probing saat collision:
  1. Hitung index = hash(key) % size
  2. Jika table[index] KOSONG → simpan di sini.
  3. Jika table[index] TERISI (collision) →
       coba slot berikutnya: index = (index + 1) % size
  4. Ulangi langkah 3 sampai slot kosong ditemukan.
  5. Jika seluruh tabel sudah penuh → raise Exception.

  Contoh konkret:
    hash("A") = ord("A") % 5 = 65 % 5 = 0  → simpan di slot 0
    hash("F") = ord("F") % 5 = 70 % 5 = 0  → collision!
      → coba slot 1 → kosong → simpan di slot 1

  ✅ Keuntungan : implementasi sederhana, cache-friendly.
  ❌ Kekurangan : "primary clustering" — rantai slot penuh
                  yang berdekatan memperlambat pencarian.
""")

    # ── Eksperimen 2: Chaining ────────────────────────────────
    print("█" * 50)
    print("  EKSPERIMEN 2: CHAINING (ukuran tabel = 3)")
    print("█" * 50)

    ct = ChainingHashTable(size=3)

    data_ct = [("apel", 100), ("melon", 200), ("lemon", 300), ("pisang", 400)]
    for key, value in data_ct:
        ct.insert(key, value)

    ct.display()

    print("""
  📝 PENJELASAN — Bagaimana Chaining menangani collision?
  ─────────────────────────────────────────────────────────────
  Setiap slot di tabel bukan menyimpan SATU nilai, melainkan
  sebuah LINKED LIST (atau list Python) berisi semua pasangan
  key-value yang memiliki hash yang sama.

  Alur insert:
  1. Hitung index = hash(key) % size
  2. Tambahkan (key, value) ke list di table[index]
     — tidak peduli apakah slot sudah berisi data lain.

  Alur search:
  1. Hitung index = hash(key) % size
  2. Telusuri list di table[index] satu per satu
     sampai key yang cocok ditemukan.

  Contoh dari eksperimen ini:
    hash("apel")  = (97+112+101+108) % 3 = 418 % 3 = 1
    hash("melon") = (109+101+108+111+110) % 3 = 539 % 3 = 2
    hash("lemon") = (108+101+109+111+110) % 3 = 539 % 3 = 2  ← collision!
    hash("pisang")= (112+105+115+97+110+103) % 3 = 642 % 3 = 0

    Slot [2] menyimpan chain: ("melon",200) → ("lemon",300)

  ✅ Keuntungan : tidak ada batas jumlah elemen per slot,
                  tidak ada clustering.
  ❌ Kekurangan : overhead memori (pointer list), kurang
                  cache-friendly dibanding linear probing.
""")


if __name__ == "__main__":
    main()
```
