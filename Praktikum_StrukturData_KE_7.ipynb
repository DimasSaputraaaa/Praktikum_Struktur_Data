## Membuat Binary Tree Secara Manual

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

# Membuat Tree Secara Manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

# Fungsi inorder traversal
def inorder (node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

print("hasil inorder travelsal dari binary Tree: ")
inorder(root)
```

## Membuat Binary Search Tree

```python
class Node:
  def __init__ (self, data) :
    self.data = data
    self.left = None
    self.right = None

class BinarySearchTree:
  def __init__(self) :
    self.root = None
  def insert (self, root, data) :
    if root is None:
      return Node (data)
    if data < root.data:
      root.left = self.insert (root.left, data)
    else:
      root.right = self.insert (root.right, data)
    return root

bst = BinarySearchTree ()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]
for value in data_list:
  root = bst.insert(root, value)
```

## Traversal pada BST

```python
def inorder (node) :
  if node:
    inorder (node.left)
    print (node.data, end=' ')
    inorder (node.right)

def preorder (node) :
  if node:
    print (node.data, end=' ')
    preorder (node.left)
    preorder (node.right)

def postorder (node) :
  if node:
    postorder (node.left)
    postorder (node.right)
    print (node.data, end=' ')

print ("Inorder Traversal:")
inorder (root)
print ("\nPreorder Traversal:")
preorder (root)
print("\nPostorder Traversal:")
postorder (root)
```

## Tugas 2: Program Binary Search Tree (BST)

```python
# =========================================
# Program Binary Search Tree (BST)
# =========================================

# Membuat node BST
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


# Membuat class BST
class BST:
    def __init__(self):
        self.root = None

    # Fungsi insert data ke BST
    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert(self.root, data)

    # Fungsi rekursif insert
    def _insert(self, current, data):

        # Jika data lebih kecil ke kiri
        if data < current.data:
            if current.left is None:
                current.left = Node(data)
            else:
                self._insert(current.left, data)

        # Jika data lebih besar ke kanan
        elif data > current.data:
            if current.right is None:
                current.right = Node(data)
            else:
                self._insert(current.right, data)

    # Traversal inorder
    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(f"{node.data:03}", end=" ")
            self.inorder(node.right)

    # Traversal preorder
    def preorder(self, node):
        if node:
            print(f"{node.data:03}", end=" ")
            self.preorder(node.left)
            self.preorder(node.right)

    # Traversal postorder
    def postorder(self, node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(f"{node.data:03}", end=" ")

    # Fungsi pencarian
    def search(self, node, key):

        # Jika node kosong
        if node is None:
            return False

        # Jika data ditemukan
        if node.data == key:
            return True

        # Cari ke kiri
        elif key < node.data:
            return self.search(node.left, key)

        # Cari ke kanan
        else:
            return self.search(node.right, key)


# =========================================
# Data dari tugas
# =========================================

# 078 ditulis 78 agar tidak error di Python
data_angka = [78, 28, 12, 68, 105, 77]

# Membuat objek BST
tree = BST()

# Memasukkan data ke BST
for angka in data_angka:
    tree.insert(angka)

# =========================================
# Menampilkan traversal
# =========================================

print("Data BST :", [f"{x:03}" for x in data_angka])

print("\nTraversal Inorder :")
tree.inorder(tree.root)

print("\n\nTraversal Preorder :")
tree.preorder(tree.root)

print("\n\nTraversal Postorder :")
tree.postorder(tree.root)

# =========================================
# Pencarian data
# =========================================

# Tiga digit terakhir NPM
cari1 = 78

# Angka yang tidak ada dalam BST
cari2 = 99

print("\n\nHasil Pencarian:")

if tree.search(tree.root, cari1):
    print(f"Angka {cari1:03} ditemukan di BST")
else:
    print(f"Angka {cari1:03} tidak ditemukan di BST")

if tree.search(tree.root, cari2):
    print(f"Angka {cari2:03} ditemukan di BST")
else:
    print(f"Angka {cari2:03} tidak ditemukan di BST")
```
